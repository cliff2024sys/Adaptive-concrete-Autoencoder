```python
# =============================================================================
# Step 1: Data Loading and Preprocessing
# =============================================================================
import pandas as pd
import numpy as np
import time, pickle
import random
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Layer, Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l1
from tensorflow.keras.callbacks import Callback

# Unzip and load the reduced CLS dataset
!tar -xvzf "./AWID-CLS-R-Trn.tar.gz"
!tar -xvzf "./AWID-CLS-R-Tst.tar.gz"

# Import CSV files without headers
train = pd.read_csv("AWID-CLS-R-Trn/1", low_memory=False)
test  = pd.read_csv("AWID-CLS-R-Tst/1", low_memory=False)

# Add headers to the datasets
train.columns = feature_names
test.columns  = feature_names

# Export datasets with headers for later use
train.to_csv("./AWID-CLS-R-Trn_with_headers.csv", index=False)
test.to_csv("./AWID-CLS-R-Tst_with_headers.csv", index=False)

# Load the header-added CSVs
train_with_header = pd.read_csv("./AWID-CLS-R-Trn_with_headers.csv", low_memory=False)
test_with_header  = pd.read_csv("./AWID-CLS-R-Tst_with_headers.csv", low_memory=False)

# Set display options for clarity
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Convert target 'class' to binary (using function)
train_with_header["class"] = train_with_header["class"].apply(changeTargetToBinary)
test_with_header["class"]  = test_with_header["class"].apply(changeTargetToBinary)

# Replace "?" with NaN and drop columns with >99.99% missing values
train_with_header.replace("?", np.nan, inplace=True)
test_with_header.replace("?", np.nan, inplace=True)
missing_values = train_with_header.isnull().sum() / len(train_with_header) * 100
cols_to_drop = missing_values[missing_values >= 99.99].index
train_with_header = train_with_header.drop(cols_to_drop, axis=1)
test_with_header  = test_with_header.drop(cols_to_drop, axis=1)

# Convert string columns to categorical (using  helper function)
string_to_category(train_with_header)
string_to_category(test_with_header)

# Fill missing values with median
fill_empty_col(train_with_header)
fill_empty_col(test_with_header)

# Normalize numerical columns and convert categorical columns to integer encoding
normalize_data(train_with_header)
normalize_data(test_with_header)
category_to_integer(train_with_header)
category_to_integer(test_with_header)

# Split data into features and target
X_train_full = train_with_header.drop("class", axis=1).values
y_train_full = train_with_header["class"].values
X_test_full  = test_with_header.drop("class", axis=1).values
y_test       = test_with_header["class"].values

print("Preprocessing complete. X_train shape:", X_train_full.shape)

# For convenience, define alternative variables (already processed)
x_train = train_with_header.drop("class", axis=1).values
x_test  = test_with_header.drop("class", axis=1).values



# =============================================================================
# Step 2: Feature Selection Methods
# =============================================================================
# 2.1 Adaptive Feature Selection using Concrete Autoencoder
# ---------------------------------------------------------------------


class ConcreteSelect(Layer):
    """
    Adaptive Feature Selection Layer using the Concrete (Gumbel-Softmax) Distribution.
    Learns a continuous mask over input features via trainable logits.
    A stronger L1 regularization is applied to enforce sparsity and the temperature is annealed.
    """
    def __init__(self, input_dim, init_temp=5.0, min_temp=0.1, l1_reg=1e-1, **kwargs):
        super().__init__(**kwargs)
        self.input_dim = input_dim
        self.init_temp = init_temp
        self.min_temp  = min_temp
        self.l1_reg    = l1_reg
        self.temperature = tf.Variable(init_temp, trainable=False, dtype=tf.float32)
        self.logits = self.add_weight(
            name="feature_logits",
            shape=(input_dim,),
            initializer=tf.keras.initializers.RandomUniform(-0.1, 0.1),
            regularizer=l1(self.l1_reg),  # Stronger L1 for more sparsity
            trainable=True
        )

    def call(self, inputs, training=None):
        probs = tf.nn.sigmoid(self.logits)
        if training:
            # Differentiable sampling using the Concrete (Gumbel-Softmax) distribution
            uniform_noise = tf.random.uniform(tf.shape(probs))
            gumbel_noise  = -tf.math.log(-tf.math.log(uniform_noise + 1e-6) + 1e-6)
            noisy_logits  = (tf.math.log(probs + 1e-6) - tf.math.log(1.0 - probs + 1e-6) + gumbel_noise) / self.temperature
            gates = tf.nn.sigmoid(noisy_logits)
        else:
            gates = tf.cast(probs > 0.5, tf.float32)
        gates = tf.broadcast_to(gates, tf.shape(inputs))
        return inputs * gates

    def get_selected_features(self, threshold=0.85):
        """
        Returns indices of features with selection probability above the given threshold.
        """
        probs = tf.nn.sigmoid(self.logits).numpy()
        return np.where(probs > threshold)[0]

class TemperatureAnnealer(Callback):
    """
    Callback to anneal the temperature of the Concrete selector during training.
    """
    def __init__(self, concrete_layer, total_epochs, verbose=0):
        super().__init__()
        self.concrete_layer = concrete_layer
        self.total_epochs = total_epochs
        self.verbose = verbose

    def on_epoch_end(self, epoch, logs=None):
        new_temp = max(
            self.concrete_layer.init_temp *
            ((self.concrete_layer.min_temp / self.concrete_layer.init_temp) ** ((epoch+1) / self.total_epochs)),
            self.concrete_layer.min_temp
        )
        self.concrete_layer.temperature.assign(new_temp)
        if self.verbose:
            print(f"Epoch {epoch+1}: Updated Concrete layer temperature to {new_temp:.4f}")

class FeatureSelectionEarlyStop(Callback):
    """
    Early stopping based on relative improvement in validation loss.
    """
    def __init__(self, threshold=0.01, patience=5, verbose=1):
        super().__init__()
        self.threshold = threshold
        self.patience  = patience
        self.wait      = 0
        self.best_loss = None
        self.verbose   = verbose

    def on_epoch_end(self, epoch, logs=None):
        val_loss = logs.get('val_loss')
        if val_loss is None:
            return
        if self.best_loss is None:
            self.best_loss = val_loss
        else:
            improvement = (self.best_loss - val_loss) / max(self.best_loss, 1e-8)
            if improvement < self.threshold:
                self.wait += 1
            else:
                self.wait = 0
                self.best_loss = val_loss
            if self.wait >= self.patience:
                if self.verbose:
                    print(f"Stopping early at epoch {epoch+1}: relative improvement {improvement:.4f} < {self.threshold}")
                self.model.stop_training = True

# Build the Concrete Autoencoder Model using all features as input
input_dim = x_train.shape[1]
inputs = Input(shape=(input_dim,))
selector_layer = ConcreteSelect(input_dim=input_dim, init_temp=5.0, min_temp=0.1, l1_reg=1e-1, name="concrete_selector")
selected_features = selector_layer(inputs)
hidden = Dense(64, activation='relu')(selected_features)
reconstructed = Dense(input_dim, activation='linear')(hidden)
autoencoder = Model(inputs, reconstructed)
autoencoder.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss='mse')

# Train the autoencoder (using x_train for both training and validation)
annealer = TemperatureAnnealer(selector_layer, total_epochs=100, verbose=1)
early_stop = FeatureSelectionEarlyStop(threshold=0.01, patience=5, verbose=1)
print("\nTraining Adaptive Concrete Autoencoder for Feature Selection...")
autoencoder.fit(
    x_train, x_train,
    epochs=100,
    batch_size=256,
    validation_data=(x_test, x_test),
    callbacks=[annealer, early_stop],
    verbose=1
)

# Extract selected feature indices from Concrete AE and print names as well
cae_selected_idx = selector_layer.get_selected_features(threshold=0.85)
print("Selected Features (indices) by ConcreteAE:", cae_selected_idx)
cae_selected_names = [feature_names [i] for i in cae_selected_idx]
print("Selected Features (names) by ConcreteAE:", cae_selected_names)

# Split train for further evaluation
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full,
                                                  test_size=0.2, random_state=42)

# 2.2 Information Gain (IG)
from sklearn.feature_selection import mutual_info_classif
mi_scores = mutual_info_classif(X_train_full, y_train_full, random_state=42)
ig_scores = sorted(list(enumerate(mi_scores)), key=lambda x: x[1], reverse=True)
top_k = 20
ig_selected_idx = [idx for idx, score in ig_scores[:top_k]]
print("IG selected feature indices (top 20):", ig_selected_idx)
ig_selected_names = [feature_names [i] for i in ig_selected_idx]
print("IG selected feature names (top 20):", ig_selected_names)

# 2.3 Principal Component Analysis (PCA)
from sklearn.decomposition import PCA
pca = PCA(n_components=20, random_state=42)
X_train_pca = pca.fit_transform(X_train_full)
X_test_pca  = pca.transform(X_test_full)
print("PCA explained variance (20 components): {:.2f}%".format(pca.explained_variance_ratio_.sum()*100))

# 2.4 Recursive Feature Elimination (RFE)
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
rf_estimator = RandomForestClassifier(n_estimators=50, random_state=42)
rfe = RFE(estimator=rf_estimator, n_features_to_select=20, step=1)
rfe.fit(X_train_full, y_train_full)
rfe_selected_idx = [i for i, selected in enumerate(rfe.support_) if selected]
print("RFE selected feature indices (20 features):", rfe_selected_idx)
rfe_selected_names = [feature_names [i] for i in rfe_selected_idx]
print("RFE selected feature names (20 features):", rfe_selected_names)

# 2.5 Genetic Algorithm (GA) Feature Selection (Simplified)
# Here we use RandomForestClassifier instead of LogisticRegression.
from sklearn.ensemble import RandomForestClassifier
def evaluate_subset(mask, X_train, y_train, X_val, y_val):
    selected_idx = [i for i, m in enumerate(mask) if m == 1]
    if len(selected_idx) == 0:
        return 0
    model = RandomForestClassifier(n_estimators=50, random_state=42)
    model.fit(X_train[:, selected_idx], y_train)
    return model.score(X_val[:, selected_idx], y_val)

num_features = X_train_full.shape[1]
population_size = 20
num_generations = 10
population = [np.random.randint(2, size=num_features) for _ in range(population_size)]
# For GA evaluation, split a portion of the training data
X_train_ga, X_val_ga, y_train_ga, y_val_ga = train_test_split(X_train_full, y_train_full, test_size=0.2, random_state=42)

for gen in range(num_generations):
    fitness_scores = [evaluate_subset(ind, X_train_ga, y_train_ga, X_val_ga, y_val_ga) for ind in population]
    best_score = max(fitness_scores)
    best_mask  = population[np.argmax(fitness_scores)]
    print(f"GA Generation {gen+1}: Best validation accuracy = {best_score:.4f}")
    sorted_pop = [x for _, x in sorted(zip(fitness_scores, population), key=lambda pair: pair[0], reverse=True)]
    survivors  = sorted_pop[:population_size // 2]
    offspring = []
    while len(offspring) < population_size - len(survivors):
        parent1, parent2 = random.sample(survivors, 2)
        crossover_point = np.random.randint(1, num_features-1)
        child = np.concatenate([parent1[:crossover_point], parent2[crossover_point:]])
        offspring.append(child)
    mutation_rate = 0.01
    for child in offspring:
        for i in range(num_features):
            if np.random.rand() < mutation_rate:
                child[i] = 1 - child[i]
    population = survivors + offspring
ga_selected_idx = [i for i, m in enumerate(best_mask) if m == 1]
print("GA selected feature indices:", ga_selected_idx)
ga_selected_names = [feature_names [i] for i in ga_selected_idx]
print("GA selected feature names:", ga_selected_names)

# Define dictionary of feature selection data
X_train_all = X_train_full.copy()
X_test_all  = X_test_full.copy()
X_train_IG  = X_train_full[:, ig_selected_idx]
X_test_IG   = X_test_full[:, ig_selected_idx]
X_train_RFE = X_train_full[:, rfe_selected_idx]
X_test_RFE  = X_test_full[:, rfe_selected_idx]
X_train_GA  = X_train_full[:, ga_selected_idx]
X_test_GA   = X_test_full[:, ga_selected_idx]
X_train_CAE = X_train_full[:, cae_selected_idx]
X_test_CAE  = X_test_full[:, cae_selected_idx]


fs_data = {
    "All":        (X_train_all, X_test_all),
    "IG":         (X_train_IG, X_test_IG),
    "RFE":        (X_train_RFE, X_test_RFE),
    "GA":         (X_train_GA, X_test_GA),
    "ConcreteAE": (X_train_CAE, X_test_CAE),
    "PCA":        (X_train_pca, X_test_pca)
}

# =============================================================================
# Step 3: Classifier Training and Single-Pass Evaluation
# =============================================================================
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
import lightgbm as lgb

# Define classifiers 
classifiers = {
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42),
    "ExtraTrees":   ExtraTreesClassifier(n_estimators=100, random_state=42),
    "LightGBM":     lgb.LGBMClassifier(n_estimators=100, random_state=42)
}

results_single = {fs_method: {} for fs_method in fs_data.keys()}

print("\nSingle-pass classifier evaluation (training on full train, testing on test):")
for fs_method, (X_tr, X_te) in fs_data.items():
    print(f"\nFS Method: {fs_method} - Feature dimension: {X_tr.shape[1]}")
    results_single[fs_method] = {}
    for clf_name, clf in classifiers.items():
        clf_instance = pickle.loads(pickle.dumps(clf))
        start_train = time.time()
        clf_instance.fit(X_tr, y_train_full)
        train_time = time.time() - start_train
        start_infer = time.time()
        y_pred = clf_instance.predict(X_te)
        infer_time = time.time() - start_infer
        acc  = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        rec  = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1   = f1_score(y_test, y_pred, average='weighted', zero_division=0)
        model_size = len(pickle.dumps(clf_instance)) / 1024
        results_single[fs_method][clf_name] = {
            "accuracy": acc,
            "precision": prec,
            "recall": rec,
            "f1_score": f1,
            "train_time_sec": train_time,
            "infer_time_sec": infer_time,
            "model_size_kb": model_size
        }
        print(f"  {clf_name}: Acc={acc:.4f}, Prec={prec:.4f}, Rec={rec:.4f}, F1={f1:.4f}, TrainTime={train_time:.2f}s, InferTime={infer_time:.4f}s, Size={model_size:.1f} KB")

# =============================================================================
# Step 4: 5-Fold Cross-Validation for All FS Methods + Classifiers
# =============================================================================
from sklearn.model_selection import StratifiedKFold

print("\nPerforming 5-fold cross-validation for each FS method + classifier...")
results_cv = []  # list to store CV metrics

def evaluate_cv_performance(clf, X_data, y_data, folds=5):
    cv = StratifiedKFold(n_splits=folds, shuffle=True, random_state=42)
    acc_list, prec_list, rec_list, f1_list = [], [], [], []
    train_times, infer_times = [], []
    fold_acc = []  # store per-fold accuracies for significance testing
    for train_idx, test_idx in cv.split(X_data, y_data):
        X_tr_fold, X_te_fold = X_data[train_idx], X_data[test_idx]
        y_tr_fold, y_te_fold = y_data[train_idx], y_data[test_idx]
        start_train = time.time()
        clf.fit(X_tr_fold, y_tr_fold)
        train_times.append(time.time() - start_train)
        start_infer = time.time()
        y_pred_fold = clf.predict(X_te_fold)
        infer_times.append(time.time() - start_infer)
        acc = accuracy_score(y_te_fold, y_pred_fold)
        fold_acc.append(acc)
        acc_list.append(acc)
        prec_list.append(precision_score(y_te_fold, y_pred_fold, average='weighted', zero_division=0))
        rec_list.append(recall_score(y_te_fold, y_pred_fold, average='weighted', zero_division=0))
        f1_list.append(f1_score(y_te_fold, y_pred_fold, average='weighted', zero_division=0))
    return {
        'cv_accuracy':  np.mean(acc_list),
        'cv_precision': np.mean(prec_list),
        'cv_recall':    np.mean(rec_list),
        'cv_f1_score':  np.mean(f1_list),
        'cv_train_time_sec': np.mean(train_times),
        'cv_infer_time_sec': np.mean(infer_times),
        'fold_accuracies': fold_acc
    }

for fs_method, (X_tr, X_te) in fs_data.items():
    print(f"\n5-fold CV on FS Method: {fs_method}, dimension={X_tr.shape[1]}")
    for clf_name, clf_proto in classifiers.items():
        clf_instance = pickle.loads(pickle.dumps(clf_proto))
        cv_metrics = evaluate_cv_performance(clf_instance, X_tr, y_train_full, folds=5)
        clf_full = pickle.loads(pickle.dumps(clf_proto))
        clf_full.fit(X_tr, y_train_full)
        model_size_kb = len(pickle.dumps(clf_full)) / 1024
        row = {
            'Feature_Selection': fs_method,
            'Classifier': clf_name,
            'CV_Accuracy': cv_metrics['cv_accuracy'],
            'CV_Precision': cv_metrics['cv_precision'],
            'CV_Recall': cv_metrics['cv_recall'],
            'CV_F1_Score': cv_metrics['cv_f1_score'],
            'CV_TrainTime': cv_metrics['cv_train_time_sec'],
            'CV_InferTime': cv_metrics['cv_infer_time_sec'],
            'Model_Size_KB': model_size_kb,
            'Fold_Accuracies': cv_metrics['fold_accuracies']
        }
        results_cv.append(row)
        print(f"  {clf_name}: CV_Acc={row['CV_Accuracy']:.4f}, CV_Precision={row['CV_Precision']:.4f}, CV_Recall={row['CV_Recall']:.4f}, CV_F1={row['CV_F1_Score']:.4f}, TrainTime={row['CV_TrainTime']:.2f}s, InferTime={row['CV_InferTime']:.4f}s, Size={model_size_kb:.1f} KB")

cv_results_df = pd.DataFrame(results_cv)
cv_results_df.to_csv("feature_selection_comparison_cv_results.csv", index=False)

# =============================================================================
# Step 5: Statistical Significance Testing (Wilcoxon Signed-Rank Test)
# =============================================================================
from scipy.stats import wilcoxon

# pairwise tests: ConcreteAE vs. others (RandomForest)
methods_to_compare = ["ConcreteAE", "IG", "RFE", "GA", "PCA"]
rf_fold_accuracies = {}
for method in methods_to_compare:
    subset = cv_results_df[(cv_results_df['Feature_Selection'] == method) & (cv_results_df['Classifier'] == "RandomForest")]
    if not subset.empty:
        rf_fold_accuracies[method] = np.array(subset.iloc[0]['Fold_Accuracies'])

print("\nStatistical Significance Testing (Wilcoxon Signed-Rank Test) comparing ConcreteAE vs. others (RandomForest):")
for method in methods_to_compare:
    if method == "ConcreteAE":
        continue
    if method in rf_fold_accuracies and "ConcreteAE" in rf_fold_accuracies:
        stat, p_value = wilcoxon(rf_fold_accuracies["ConcreteAE"], rf_fold_accuracies[method])
        sig = "Significant" if p_value < 0.05 else "Not significant"
        print(f"ConcreteAE vs {method}: p-value = {p_value:.4f} ({sig})")

# --- Extended Comparisons ---
# Now include additional pairwise comparisons and compare each method with baseline (All features)
fs_methods_extended = ["All", "ConcreteAE", "IG", "RFE", "GA", "PCA"]
rf_fold_accuracies_extended = {}
for method in fs_methods_extended:
    subset = cv_results_df[(cv_results_df['Feature_Selection'] == method) & (cv_results_df['Classifier'] == "RandomForest")]
    if not subset.empty:
        rf_fold_accuracies_extended[method] = np.array(subset.iloc[0]['Fold_Accuracies'])

print("\nExtended Statistical Significance Testing (Wilcoxon Signed-Rank Test):")
#  pairwise comparisons
comparisons = [
    ("IG", "GA"),
    ("GA", "RFE"),
    ("ConcreteAE", "RFE"),
    ("GA", "PCA")
]
for m1, m2 in comparisons:
    if m1 in rf_fold_accuracies_extended and m2 in rf_fold_accuracies_extended:
        stat, p_value = wilcoxon(rf_fold_accuracies_extended[m1], rf_fold_accuracies_extended[m2])
        sig = "Significant" if p_value < 0.05 else "Not significant"
        print(f"{m1} vs {m2}: p-value = {p_value:.4f} ({sig})")

# Compare each FS method (excluding All) against the baseline (All features)
for method in ["ConcreteAE", "IG", "RFE", "GA", "PCA"]:
    if "All" in rf_fold_accuracies_extended and method in rf_fold_accuracies_extended:
        stat, p_value = wilcoxon(rf_fold_accuracies_extended["All"], rf_fold_accuracies_extended[method])
        sig = "Significant" if p_value < 0.05 else "Not significant"
        print(f"All Features vs {method}: p-value = {p_value:.4f} ({sig})")

# =============================================================================
# Step 6: Summarize and Save Final Results
# =============================================================================
# 6A. Single-run Results with extended evaluation metrics (including precision and recall)
records_single = []
for fs_method, clf_dict in results_single.items():
    for clf_name, metrics in clf_dict.items():
        rec = {"FS_Method": fs_method, "Classifier": clf_name}
        rec.update(metrics)
        records_single.append(rec)
single_df = pd.DataFrame(records_single)
print("\n=== Single-Run Results ===")
print(single_df[['FS_Method','Classifier','accuracy','precision','recall','f1_score','train_time_sec','infer_time_sec','model_size_kb']])
single_df.to_csv("feature_selection_comparison_single_run.csv", index=False)

# 6B. Cross-Validation Results with all evaluation metrics
print("\n=== 5-Fold Cross-Validation Results ===")
print(cv_results_df[['Feature_Selection','Classifier','CV_Accuracy','CV_Precision','CV_Recall','CV_F1_Score','CV_TrainTime','CV_InferTime','Model_Size_KB']].head(20))
cv_results_df.to_csv("feature_selection_comparison_cv_results.csv", index=False)

print("\nAll done! Single-run and cross-validation results are saved.")

# =============================================================================
# Step 7: Plotting Evaluation Metrics and Computational Efficiency
# =============================================================================
# 7A. Generate a grouped bar plot for evaluation metrics (accuracy, precision, recall, F1 score)
metrics_agg = cv_results_df.groupby('Feature_Selection').agg({
    'CV_Accuracy': 'mean',
    'CV_Precision': 'mean',
    'CV_Recall': 'mean',
    'CV_F1_Score': 'mean'
}).reset_index()

x = np.arange(len(metrics_agg['Feature_Selection']))
width = 0.2

fig, ax = plt.subplots(figsize=(10,6))
ax.bar(x - 1.5*width, metrics_agg['CV_Accuracy'], width, label='Accuracy')
ax.bar(x - 0.5*width, metrics_agg['CV_Precision'], width, label='Precision')
ax.bar(x + 0.5*width, metrics_agg['CV_Recall'], width, label='Recall')
ax.bar(x + 1.5*width, metrics_agg['CV_F1_Score'], width, label='F1 Score')

ax.set_ylabel('Scores')
ax.set_title('Evaluation Metrics by Feature Selection Method (5-Fold CV)')
ax.set_xticks(x)
ax.set_xticklabels(metrics_agg['Feature_Selection'])
ax.legend()
plt.tight_layout()
plt.show()

# 7B. Plot computational efficiency (average training time) for each feature selection method
time_agg = cv_results_df.groupby('Feature_Selection').agg({
    'CV_TrainTime': 'mean'
}).reset_index()

fig, ax = plt.subplots(figsize=(8,5))
ax.bar(time_agg['Feature_Selection'], time_agg['CV_TrainTime'])
ax.set_ylabel('Average Training Time (sec)')
ax.set_title('Computational Efficiency (Training Time) by Feature Selection Method')
plt.tight_layout()
plt.show()
```
